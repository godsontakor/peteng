```julia
x = collect(linspace(0, 10, 100))
y = sin.(x)
z = cos.(x)
plot(x, y, "--r", x, z, ":g")
xlabel("x axis")
ylabel("y axis")
title("sin and cos plots")
grid()
text(6.0, 1.0, "annotation")
savefig("sincos.eps")
```
